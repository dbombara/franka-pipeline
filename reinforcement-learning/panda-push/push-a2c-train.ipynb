```python
import gymnasium as gym
import numpy as np
import panda_gym
from stable_baselines3 import A2C
from stable_baselines3.common.env_util import make_vec_env
import PIL
from PIL import Image, ImageDraw
```

```python
# Create the environment
env = make_vec_env('PandaPush-v3',n_envs=4, seed=98)
```

```python
model = A2C('MultiInputPolicy', 
            env, 
            verbose = 1,
            ent_coef = 0.0,
            max_grad_norm = 0.5,
            n_steps = 8,
            gae_lambda = 0.9,
            vf_coef = 0.4,
            gamma = 0.9,
            use_rms_prop = True,
            normalize_advantage = False,
            learning_rate = 7e-4,
            use_sde = True)
```

```python
# Train for the specified timesteps 
model.learn(total_timesteps = 1e6)
```

```python
# Save the trained agent
model.save('agent/a2c_pandapush_agent')
policy = model.policy
policy.save("policy/a2c_pandapush_policy")
```
